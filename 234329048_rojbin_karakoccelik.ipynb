## QUESTION 4

```python
freq_table = pd.crosstab(df["Drug_Name"], df["Side_Effects"])
freq_table
```

```python
plt.figure(figsize=(16, 10))
sns.heatmap(freq_table, cmap="viridis")
plt.title("Drug_Name vs Side_Effects - 2D Frequency Heatmap")
plt.xlabel("Side Effects")
plt.ylabel("Drug Name")
plt.tight_layout()
plt.show()
```

## QUESTION 5

```python
plt.figure(figsize=(12, 12))

sns.clustermap(
    freq_table,
    cmap="viridis",
    figsize=(14, 14),
    metric="euclidean",
    method="ward",
    linewidths=0
)

plt.title("Clustermap of Drug_Name vs Side_Effects", pad=100)
plt.show()
```

The clustermap visualizes the joint frequency structure between Drug_Name and Side_Effects using hierarchical clustering. The heatmap shows that most drugside-effect combinations occur with very low frequency, while a small set of drugs exhibit clearly defined high-frequency side-effect patterns.

On the row dendrogram, drugs such as Tramadol, Metformin, Losartan, Ibuprofen, and Escitalopram form localized clusters, indicating that they share similar side-effect profiles. Similarly, the column dendrogram groups common side effects such as Dizziness, Nausea, Injection-site pain, which appear across multiple drugs and therefore cluster tightly together. Rare side effects form broader, sparse clusters due to their low occurrence.

Overall, the clustermap reveals meaningful structure:

Drugs with comparable adverse-effect patterns cluster together,

Frequently reported side effects group into shared clusters,

Rare effects distribute into diffuse, low-density regions.

This hierarchical organization confirms that the dataset exhibits coherent and interpretable drugside-effect relationships.

## QUESTION 6

```python
corr_pearson = df[["Age", "Improvement_Score"]].corr(method="pearson")
print("Pearson correlation matrix:")
print(corr_pearson)

# stersen Spearman da bakabilirsin:
corr_spearman = df[["Age", "Improvement_Score"]].corr(method="spearman")
print("\nSpearman correlation matrix:")
print(corr_spearman)
```

```python
plt.figure(figsize=(7, 5))
sns.regplot(
    data=df,
    x="Age",
    y="Improvement_Score",
    scatter_kws={"alpha": 0.6}
)
plt.title("Relationship between Age and Improvement Score")
plt.xlabel("Age")
plt.ylabel("Improvement Score")
plt.tight_layout()
plt.show()
```

Both the Pearson (0.0105) and Spearman (0.0117) correlation coefficients indicate an extremely weak positive association between Age and Improvement_Score. These values are very close to zero, suggesting that there is no meaningful linear or monotonic relationship between the two variables.

The scatter plot supports this conclusion: data points are widely dispersed across all age ranges, and the regression line is nearly flat. This indicates that patients' improvement levels are essentially independent of age, and age does not appear to be a predictive factor for treatment improvement in this dataset.

In summary, the analysis provides strong evidence that Age has no significant influence on Improvement_Score.

```python
normalized = (
    df.groupby("Gender")["Condition"]
      .value_counts(normalize=True)
      .rename("Probability")
      .reset_index()
)

print(normalized)
```
